```python
cf.dropna(inplace=True)
```

```python
cf.drop_duplicates(inplace=True)
```

```python
cf
```

```python
cf.drop(columns={'Not_Useful_Column'},inplace=True)
```

```python
cf['Last_Name']=cf['Last_Name'].str.strip().str.replace(r'[_]','',regex=True)
```

```python
cf['Phone_Number']=cf['Phone_Number'].str.strip().str.replace(r'[_/|-]','',regex=True)
```

```python
cf=cf[cf['Phone_Number']!='Na']
```

```python
cf
```

```python
cf.loc[:, 'Paying Customer'] = cf['Paying Customer'].replace({'Yes': 'Yes', 'No': 'No', 'Y': 'Yes', 'N': 'No'})
```

```python
cf.loc[:,'Do_Not_Contact']=cf['Do_Not_Contact'].replace({'Yes':'Yes','No':'No','Y':'Yes','N':'No'})
```

```python
cf
```

```python
cf=cf[cf['Do_Not_Contact']!='Yes']
```
